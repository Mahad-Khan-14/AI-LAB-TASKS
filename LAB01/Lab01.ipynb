```python
# Q3

n = int(input("Enter how many numbers you want to store: "))

numbers = []

for i in range(n):
    num = int(input(f"Enter number {i + 1}: "))
    numbers.append(num)

print("\nList of numbers:")
for num in numbers:
    print(num)

even_count = 0
odd_count = 0

for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("\nEven numbers count:", even_count)
print("Odd numbers count:", odd_count)

index = int(input("\nEnter the index you want to replace (0-based): "))
new_value = int(input("Enter the new value: "))

if 0 <= index < n:
    numbers[index] = new_value
    print("\nUpdated list:", numbers)
else:
    print("\nInvalid index!")
```

```python
# Q4

def calculate_total_salary(basic_salary):
    hra = 0.10 * basic_salary   # 10% HRA
    da = 0.05 * basic_salary    # 5% DA
    total_salary = basic_salary + hra + da
    return total_salary

salary = float(input("Enter basic salary: "))

result = calculate_total_salary(salary)

print("Total Salary:", result)
```

```python
# Q5

def string_Func(text):
    result = {
        "Total Characters": len(text),
        "Uppercase Letters": 0,
        "Lowercase Letters": 0,
        "Digits": 0
    }

    for char in text:
        if char.isupper():
            result["Uppercase Letters"] += 1
        elif char.islower():
            result["Lowercase Letters"] += 1
        elif char.isdigit():
            result["Digits"] += 1

    return result

user_input = input("Enter a string: ")

analysis = string_Func(user_input)

print("\nString Analysis:")
for key, value in analysis.items():
    print(f"{key}: {value}")
```

```python
# Q6

text = input("Enter a string: ")

current_char = text[0]
current_count = 1

longest_char = text[0]
longest_count = 1

for i in range(1, len(text)):
    if text[i] == current_char:
        current_count += 1
    else:
        current_char = text[i]
        current_count = 1

    if current_count > longest_count:
        longest_count = current_count
        longest_char = current_char

longest_substring = longest_char * longest_count

print("Longest substring:", longest_substring)
print("Length:", longest_count)
```

```python
# Q7

def find_repeated_numbers(nums):
    repeated = []

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                already_added = False
                for r in repeated:
                    if r == nums[i]:
                        already_added = True
                        break
                if not already_added:
                    repeated.append(nums[i])
                break

    return repeated

n = int(input("Enter how many numbers you want to store: "))

numbers = []

for i in range(n):
    num = int(input(f"Enter number {i + 1}: "))
    numbers.append(num)

print("Repeated numbers:", find_repeated_numbers(numbers))
```
