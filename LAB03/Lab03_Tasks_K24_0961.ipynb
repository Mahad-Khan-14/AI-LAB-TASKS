```python
# Maze Representation
maze = [
['S', 0, 0, 1, 0],
[1, 0, 1, 0, 0],
[0, 0, 0, 0, 'G'],
[1, 1, 0, 1, 1]
]

rows = len(maze)
cols = len(maze[0])

# Find start position
def find_start():
    for i in range(rows):
        for j in range(cols):
            if maze[i][j] == 'S':
                return (i, j)

# Check valid move
def valid(x, y, visited):
    if x < 0 or x >= rows or y < 0 or y >= cols:
        return False
    if maze[x][y] == 1:
        return False
    if (x, y) in visited:
        return False
    return True

# DFS with depth limit
def depth_limited_dfs(x, y, limit, visited, path):

    if maze[x][y] == 'G':
        return path

    if limit == 0:
        return None

    visited.append((x, y))

    moves = [(-1,0),(1,0),(0,-1),(0,1)]

    for dx, dy in moves:
        nx = x + dx
        ny = y + dy

        if valid(nx, ny, visited):
            result = depth_limited_dfs(nx, ny, limit-1, visited, path + [(nx, ny)])

            if result:
                return result

    visited.remove((x, y))
    return None


# Iterative Deepening Search
def IDS(start, max_depth):

    for depth in range(max_depth + 1):

        visited = []
        path = depth_limited_dfs(start[0], start[1], depth, visited, [start])

        if path:
            print("Goal found at depth:", depth)
            return path

    return None


# -----------------------
# Testing
# -----------------------

start = find_start()

print("Start Position:", start)

print("\nDepth Limited DFS:")
dfs_path = depth_limited_dfs(start[0], start[1], 10, [], [start])

if dfs_path:
    print("Path to goal:", dfs_path)
else:
    print("Goal not found within limit")

print("\nIterative Deepening Search:")

ids_path = IDS(start, 10)

if ids_path:
    print("Path to goal:", ids_path)
else:
    print("Goal not found")
```
