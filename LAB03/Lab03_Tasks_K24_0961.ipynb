TASK 01

```python
import random

class Environment:
    def __init__(self):
        self.components = {}
        for comp in ['A','B','C','D','E','F','G','H','I']:
            self.components[comp] = random.choice(['Safe', 'Vulnerable'])

    def display(self):
        print("\nSystem State:")
        for k, v in self.components.items():
            print(f"{k}: {v}")

class SecurityAgent:
    def __init__(self):
        self.vulnerable_list = []

    def scan(self, env):
        print("\nScanning System...")
        for comp, state in env.components.items():
            if state == 'Vulnerable':
                print(f"{comp} is Vulnerable ")
                self.vulnerable_list.append(comp)
            else:
                print(f"{comp} is Safe ")

    def patch(self, env):
        print("\nPatching Vulnerabilities...")
        for comp in self.vulnerable_list:
            env.components[comp] = 'Safe'
            print(f"{comp} patched ")

env = Environment()
agent = SecurityAgent()

env.display()
agent.scan(env)
agent.patch(env)
env.display()
```

TASK 02

```python
import random

servers = []
states = ['Underloaded', 'Balanced', 'Overloaded']

# Initialize 5 servers
for i in range(5):
    servers.append(random.choice(states))

print("Initial Server Load:")
for i, s in enumerate(servers):
    print(f"Server {i}: {s}")

# Balancing
for i in range(len(servers)):
    if servers[i] == 'Overloaded':
        for j in range(len(servers)):
            if servers[j] == 'Underloaded':
                servers[i] = 'Balanced'
                servers[j] = 'Balanced'
                print(f"\nMoved load from Server {i} → Server {j}")
                break

print("\nFinal Server Load:")
for i, s in enumerate(servers):
    print(f"Server {i}: {s}")
```

TASK 03

```python
import random

tasks = []
for i in range(6):
    tasks.append(random.choice(['Completed', 'Failed']))

print("Initial Backup Status:")
print(tasks)

# Retry failed backups
for i in range(len(tasks)):
    if tasks[i] == 'Failed':
        print(f"Retrying Task {i}...")
        tasks[i] = 'Completed'

print("\nFinal Backup Status:")
print(tasks)
```

```python
import random

class Environment:
    def __init__(self):
        states = ['Safe', 'Low Risk', 'High Risk']
        self.components = {c: random.choice(states) for c in 'ABCDEFGHI'}

    def display(self):
        print("\nSystem State:")
        for k, v in self.components.items():
            print(f"{k}: {v}")

class UtilityAgent:
    def scan_and_patch(self, env):
        print("\nScanning System...")
        for comp, state in env.components.items():
            if state == 'Low Risk':
                print(f"{comp}: Low Risk → Patching ")
                env.components[comp] = 'Safe'
            elif state == 'High Risk':
                print(f"{comp}: High Risk → Premium Required ")
            else:
                print(f"{comp}: Safe ")

env = Environment()
agent = UtilityAgent()

env.display()
agent.scan_and_patch(env)
env.display()
```

TASK 05

```python
class HospitalAgent:
    def __init__(self):
        self.goal = "Deliver Medicine"

    def act(self, task):
        room = task['room']
        medicine = task['medicine']
        
        print(f"\nMoving to Room {room}")
        print("Picking medicine:", medicine)
        print("Scanning patient ID...")
        print("Delivering medicine ")

        if task.get('critical'):
            print("Alerting staff ")

# Environment (tasks)
tasks = [
    {'room': 101, 'medicine': 'Paracetamol'},
    {'room': 102, 'medicine': 'Insulin', 'critical': True},
    {'room': 103, 'medicine': 'Antibiotic'}
]

agent = HospitalAgent()

for t in tasks:
    agent.act(t)
```

```python
grid = {
    'a': 'safe', 'b': 'safe', 'c': 'fire',
    'd': 'safe', 'e': 'fire', 'f': 'safe',
    'g': 'safe', 'h': 'safe', 'j': 'fire'
}

path = ['a','b','c','d','e','f','g','h','j']

def display(grid):
    print("\nGrid State:")
    keys = list(grid.keys())
    for i in range(0, 9, 3):
        row = keys[i:i+3]
        for r in row:
            if grid[r] == 'fire':
                print("f", end=" ")
            else:
                print(".", end=" ")
        print()

for room in path:
    print(f"\nRobot at {room}")
    
    if grid[room] == 'fire':
        print("Fire detected! Extinguishing... ")
        grid[room] = 'safe'
    else:
        print("No fire.")

    display(grid)

print("\nFinal State (All Safe):")
display(grid)
```
