TASK 01:

```python

class SecurityAgent:
    def __init__(self, agent_id, name, status="Inactive"):
        self.agent_id = agent_id
        self.name = name
        self.status = status

    def activate(self):
        self.status = "Active"
        print(f"[INFO] Agent {self.name} activated.")

    def deactivate(self):
        self.status = "Inactive"
        print(f"[INFO] Agent {self.name} deactivated.")

    def display_info(self):
        print("---------------")
        print(f"Agent ID   : {self.agent_id}")
        print(f"Name       : {self.name}")
        print(f"Status     : {self.status}")
        print("---------------")



class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        if self.status == "Active":
            print(f"[FIREWALL] {self.name} is monitoring network traffic...")
            print("[FIREWALL] Suspicious IP detected. Blocking connection.")
        else:
            print(f"[WARNING] {self.name} is not active.")



class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        if self.status == "Active":
            print(f"[MALWARE SCANNER] {self.name} is scanning files...")
            print("[MALWARE SCANNER] Malware detected! Quarantining file.")
        else:
            print(f"[WARNING] {self.name} is not active.")



class AutomationAgent(SecurityAgent):
    def run_automation(self):
        if self.status == "Active":
            print(f"[AUTOMATION] {self.name} is running AI-based security optimization...")
            print("[AUTOMATION] Updating firewall rules and system patches automatically.")
        else:
            print(f"[WARNING] {self.name} is not active.")


firewall = FirewallAgent(101, "Network Firewall Agent")
malware_scanner = MalwareDetectionAgent(102, "File Protection Agent")
automation = AutomationAgent(103, "AI Automation Agent")

firewall.activate()
malware_scanner.activate()
automation.activate()

print("\n=== CYBERSECURITY SYSTEM RUNNING ===\n")

firewall.monitor_traffic()
print()
malware_scanner.scan_files()
print()
automation.run_automation()
```

```python

class Threat:
    def __init__(self, threat_id, name, severity):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity

    def display_info(self):
        print("---------------")
        print(f"Threat ID : {self.threat_id}")
        print(f"Name      : {self.name}")
        print(f"Severity  : {self.severity}")
        print("---------------")

    def respond(self):
        print(f"[RESPONSE] Taking automated action against {self.name}...\n")



class PhishingThreat(Threat):
    def analyze_email(self):
        print(f"[PHISHING] Analyzing suspicious email patterns...")
        print("[PHISHING] Malicious link detected.")
        self.respond()



class RansomwareThreat(Threat):
    def scan_files(self):
        print(f"[RANSOMWARE] Scanning file system for encryption activity...")
        print("[RANSOMWARE] Unauthorized encryption detected.")
        self.respond()



class BotnetThreat(Threat):
    def detect_traffic(self):
        print(f"[BOTNET] Monitoring unusual network traffic...")
        print("[BOTNET] Command-and-control traffic detected.")
        self.respond()


phishing = PhishingThreat(201, "Email Phishing Attack", "High")
ransomware = RansomwareThreat(202, "File Encryption Ransomware", "Critical")
botnet = BotnetThreat(203, "Distributed Botnet Activity", "Medium")

print("=== AI-BASED THREAT INTELLIGENCE SYSTEM ===\n")

phishing.display_info()
phishing.analyze_email()

ransomware.display_info()
ransomware.scan_files()

botnet.display_info()
botnet.detect_traffic()
```

```python

class ResponseAgent:
    def __init__(self, agent_id, name):
        self.agent_id = agent_id
        self.name = name

    def execute_response(self):
        print(f"[GENERIC RESPONSE] {self.name} is executing a response.")


class AlertAgent(ResponseAgent):
    def execute_response(self):  
        print(f"[ALERT AGENT] {self.name} is sending security notifications to administrators.")


class BlockAgent(ResponseAgent):
    def execute_response(self):  
        print(f"[BLOCK AGENT] {self.name} is blocking malicious IP addresses and processes.")


class RecoverAgent(ResponseAgent):
    def execute_response(self):  
        print(f"[RECOVER AGENT] {self.name} is restoring affected systems from secure backups.")


agent1 = AlertAgent(301, "Email Alert System")
agent2 = BlockAgent(302, "Firewall Blocker")
agent3 = RecoverAgent(303, "System Recovery Module")

agents = [agent1, agent2, agent3]

print("=== AI CYBERSECURITY RESPONSE SYSTEM ===\n")

for agent in agents:
    agent.execute_response()
```

```python
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def work(self):
        print(f"{self.name} is doing general Employee work")

class Manager(Employee):
    def work(self):
        print(f"Manager {self.name} is managing team's work")

class Developer(Employee):
    def work(self):
        print(f"Developer {self.name} is Developing a website")

class Designer(Employee):
    def work(self):
        print(f"Designer {self.name} is designing a new logo for the company website")


emp1 = Manager("Mahad", 1)
emp2 = Developer("Saad", 2)
emp3 = Designer("Ahad", 3)

employees = [emp1, emp2, emp3]

for emp in employees:
    emp.work()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = 0      

    def set_grade(self, grade):
        if 0 <= grade <= 100:     
            self.__grade = grade
            print(f"[INFO] Grade updated successfully for {self.name}.")
        else:
            print("[ERROR] Grade must be between 0 and 100.")

    def get_grade(self):
        return self.__grade

    def display_info(self):
        print(f"Student Name : {self.name}")
        print(f"Grade        : {self.__grade}")


student1 = Student("Ali")
student2 = Student("Sara")

student1.set_grade(85)
student2.set_grade(92)

print(f"{student1.name}'s Grade:", student1.get_grade())
print(f"{student2.name}'s Grade:", student2.get_grade())

student1.display_info()
student2.display_info()
```
