TASK 01(A)

```python
# Goal-Based Agent using Depth-Limited Search (DLS)

class GoalBasedAgent:
    def __init__(self, graph, goal, depth_limit):
        self.graph = graph
        self.goal = goal
        self.limit = depth_limit

    def dls(self, node, depth, path):
        # Goal Test
        if node == self.goal:
            return path

        # Depth limit check
        if depth == 0:
            return None

        # Explore neighbors
        for neighbor in self.graph[node]:
            result = self.dls(neighbor, depth - 1, path + [neighbor])
            if result:
                return result

        return None

    def act(self, start):
        return self.dls(start, self.limit, [start])


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

agent = GoalBasedAgent(graph, goal='F', depth_limit=2)
print("DLS Path:", agent.act('A'))
```

TASK 01(B)

```python
# Utility-Based Agent using Uniform Cost Search (UCS)

import heapq

class UtilityBasedAgent:
    def __init__(self, graph, goal):
        self.graph = graph
        self.goal = goal

    def ucs(self, start):
        pq = [(0, start, [start])]  # (cost, node, path)
        visited = set()

        while pq:
            cost, node, path = heapq.heappop(pq)

            if node == self.goal:
                return path, cost

            if node not in visited:
                visited.add(node)

                for neighbor, weight in self.graph[node]:
                    heapq.heappush(pq,
                                   (cost + weight, neighbor, path + [neighbor]))

        return None

    def act(self, start):
        return self.ucs(start)


# Weighted Graph
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}

agent = UtilityBasedAgent(graph, goal='F')
path, cost = agent.act('A')

print("UCS Path:", path)
print("Total Cost:", cost)
```

```python
# Traveling Salesman Problem (TSP)

import itertools

def tsp(graph, start):
    cities = list(graph.keys())
    cities.remove(start)

    min_path = None
    min_cost = float('inf')

    # Try all permutations
    for perm in itertools.permutations(cities):
        current_cost = 0
        current_city = start

        for city in perm:
            current_cost += graph[current_city][city]
            current_city = city

        # Return to start
        current_cost += graph[current_city][start]

        if current_cost < min_cost:
            min_cost = current_cost
            min_path = (start,) + perm + (start,)

    return min_path, min_cost


# Example Cost Matrix
graph = {
    'A': {'A':0, 'B':10, 'C':15, 'D':20},
    'B': {'A':10, 'B':0, 'C':35, 'D':25},
    'C': {'A':15, 'B':35, 'C':0, 'D':30},
    'D': {'A':20, 'B':25, 'C':30, 'D':0}
}

path, cost = tsp(graph, 'A')

print("Optimal Route:", path)
print("Minimum Cost:", cost)
```

TASK 03(A)

```python
# Iterative Deepening DFS (Graph)

def dls(graph, node, goal, depth, path):
    if node == goal:
        return path

    if depth == 0:
        return None

    for neighbor in graph[node]:
        result = dls(graph, neighbor, goal, depth - 1, path + [neighbor])
        if result:
            return result

    return None


def iddfs(graph, start, goal, max_depth):
    for depth in range(max_depth + 1):
        result = dls(graph, start, goal, depth, [start])
        if result:
            return result
    return None


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

print("IDDFS Path:", iddfs(graph, 'A', 'F', 5))
```

TASK 03(B)

```python
# Iterative Deepening DFS (Tree)

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []


def dls_tree(node, goal, depth, path):
    if node.value == goal:
        return path

    if depth == 0:
        return None

    for child in node.children:
        result = dls_tree(child, goal, depth - 1, path + [child.value])
        if result:
            return result

    return None


def iddfs_tree(root, goal, max_depth):
    for depth in range(max_depth + 1):
        result = dls_tree(root, goal, depth, [root.value])
        if result:
            return result
    return None


# Build Tree
root = TreeNode('A')
b = TreeNode('B')
c = TreeNode('C')
d = TreeNode('D')
e = TreeNode('E')

root.children = [b, c]
b.children = [d, e]

print("IDDFS Tree Path:", iddfs_tree(root, 'E', 5))
```
